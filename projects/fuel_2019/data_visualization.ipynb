```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price_by_state_df = pd.read_csv('datasets/price_by_state.csv', sep=';', index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(price_by_state_df, linewidths=.05, linecolor='gray', cmap="mako_r", annot=True, annot_kws={"fontsize":12}, 
            ax=ax, fmt = '.2f', center = 0)
title = plt.figtext(.5,.9,"Preo mdio do combustvel por tipo e estado (Em R$) - 1 Sem 2019", fontsize=14, ha="center")
title.set_position([.5, 0.92])
ax.xaxis.set_label_text("Tipo de combustvel", fontdict={'size':14})
ax.yaxis.set_label_text("Estado", fontdict={'size':14})
```

```python
fig.savefig('images/fuel2019.jpg')
```

```python
alegrete_fuel = pd.read_csv('datasets/alegrete_fuel.csv', sep=';', parse_dates=True, squeeze=True, index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
plt.plot(alegrete_fuel, color='purple')
plt.grid(True)
plt.xlabel('Data')
title = plt.figtext(.5,.9,"Preo mdio do combustvel em Alegrete - RS", fontsize=14, ha="center")
ax.xaxis.set_label_text("Data de Coleta", fontdict={'size':14})
ax.yaxis.set_label_text("Preo Mdio (em R$)", fontdict={'size':14})
plt.show()
```

```python
fig.savefig('images/alegrete_fuel.jpg')
```
